## merge both transaction and identity data

# sort dataframe on basis of time

```python
train_df = train_df.sort_values("TransactionDT")
train_df
```

```python
train_df.to_csv("/Users/sumanyadav/Desktop/PROJECT/DriftGuard/data/merged/merged-train-df.csv", index=False)
print("saved successfully")
```

# Creating time slice

```python
train_df["time_slice"] = pd.qcut(train_df["TransactionDT"], q = 8, labels=False)
```

```python
train_df.shape
```

```python
train_df["time_slice"].value_counts().sort_index()
```

```python
for idx in range(len(train_df["time_slice"].unique())):
    slice_df = train_df[train_df["time_slice"] == idx]
    if idx == 0:
        slice_df.to_csv("/Users/sumanyadav/Desktop/PROJECT/DriftGuard/data/time_slices/baseline-data.csv", index=False)
    else:
        slice_df.to_csv(f"/Users/sumanyadav/Desktop/PROJECT/DriftGuard/data/time_slices/slice-{idx}.csv", index=False)
print("saved successfully")
```

```python
missing = baseline.isna().mean()
missing.to_json("missing.json", indent=2)
```

```python
baseline.shape
```

```python
numeric_cols = baseline.select_dtypes(include=["int64", "float64"]).columns
categorical_cols = baseline.select_dtypes(include=["object", "string"]).columns
ignore_cols = ["TransactionID", "isFraud", "time_slice"]

numeric_cols.size, categorical_cols.size
```

```python
numeric_cols = [col for col in numeric_cols if col not in ignore_cols]
outlier_bounds = {}

for col in numeric_cols:
    q1 = baseline[col].quantile(0.25)
    q3 = baseline[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5*iqr
    upper_bound = q1 + 1.5*iqr

    outlier_bounds[col] = {
        "Q1": q1,
        "Q3": q3,
        "IQR": iqr,
        "Lower Bound": lower_bound,
        "Upper Bound": upper_bound
    }
```

```python
baseline.describe().to_json("describe_baseline.json", indent=2)
```

```python
import json
with open("outlier_bounds.json", "w") as f:
    json.dump(outlier_bounds, f, indent=2)
```
